# **Loading and preprocessing the dataset**

```python
# Loading the dataset
import pandas as pd
df_sensors = pd.read_csv("sensor.csv")
df_sensors.head()
```

```python
# Here is the dataset each coulumn description
df_sensors.describe().T
```

```python
# As many of the records contains empty values in it so we delete that records
df_sensors.dropna(inplace=True)
```

```python
# Seperating the features that are from sensor_00 to sensor_51
# and the target column as y that is the machine status column
X = df_sensors.iloc[:, 2:-1]
y = df_sensors.iloc[:,-1]
```

```python
# Converting the label of the target column into numbers
# 0 -> NORMAL
# 1 -> RECOVERING
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Here is the preprocessed features and label columns shape
print("features shape:", X.shape)
print("labels shape:", y.shape)      
```

# **Applying the outlier Model**

```python
# Now we are preparing the features data for the outlier detection
import numpy as np # For arrays manupulation
from alibi_detect.utils.data import create_outlier_batch # for creating the outlier batch

# Initializing the random seed to zero 
np.random.seed(0) 

# Creating the batch of the features and labels into 110000 samples
normal_batch = create_outlier_batch(X.values, y, n_samples=110000, perc_outlier=0)

# Making the features values to floating poin values
X_train, y_train = normal_batch.data.astype('float'), normal_batch.target

# Normalizing the train data
mean, stdev = X_train.mean(axis=0), X_train.std(axis=0)
X_train = (X_train - mean) / stdev

print("Outlier data shape:", X_train.shape)
```

```python
# initialize outlier detector
from alibi_detect.od import IForest

# Train the data outlier detector
od = IForest(threshold=None,n_estimators=100)
od.fit(X_train)
```

```python
# create batch of outlier
np.random.seed(0)
perc_outlier = 5

# Creating the batch of the features and labels into 1000 samples
threshold_batch = create_outlier_batch(X.values, y, n_samples=1000, perc_outlier=perc_outlier)
X_threshold, y_threshold = threshold_batch.data.astype('float'), threshold_batch.target
X_threshold = (X_threshold - mean) / stdev
print('{}% outliers'.format(100 * y_threshold.mean()))

# Add threshold to detector
od.infer_threshold(X_threshold, threshold_perc=100-perc_outlier)
print('New threshold: {}'.format(od.threshold))
```

```python
# new batch for prediction
np.random.seed(1)
outlier_batch = create_outlier_batch(X.values, y, n_samples=1000, perc_outlier=10)
X_outlier, y_outlier = outlier_batch.data.astype('float'), outlier_batch.target
X_outlier = (X_outlier - mean) / stdev
print(X_outlier.shape, y_outlier.shape)
print('{}% outliers'.format(100 * y_outlier.mean()))
```

```python
# Saving the model 
import pickle
pickle.dump(od, open("Model", 'wb'))
```

# **Evaluating the results**

```python
# Loading the model
loaded_model = pickle.load(open("Model", 'rb'))
```

```python
# predicting the outliers from the loaded model
od_preds = loaded_model.predict(X_outlier, return_instance_score=True)
```

```python
# Here are we comparing the actual values and the predicted values using a confusion matrics, F! score and Accuracy Score
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score
import seaborn as sns

# Here is the code to for the Accuracy score
labels = le.classes_
y_pred = od_preds['data']['is_outlier']
f1 = f1_score(y_outlier, y_pred)
acc = accuracy_score(y_outlier, y_pred)
print('F1 score: {:.4f},\nAccuracy Score: {:.4f}'.format(f1, acc))

# Here is the code for the confusion matrics
cm = confusion_matrix(y_outlier, y_pred)
df_cm = pd.DataFrame(cm, index=labels, columns=labels)
sns.heatmap(df_cm, annot=True, cbar=True, linewidths=.5)
```

```python
# Here are we dispalying the 1000 values of the outliear and the normal values that are under the threshold
from alibi_detect.utils.visualize import plot_instance_score
plot_instance_score(od_preds, y_outlier, labels, od.threshold)
```
