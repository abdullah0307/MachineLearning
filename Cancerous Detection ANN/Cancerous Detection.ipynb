**Importing Dataset**

```python
import pandas as pd
dataset = pd.read_excel("clinical_dataset.xlsx")
```

**Each row information**

```python
dataset.info()
```

**Preprocess the dataset**

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
y = le.fit_transform(dataset['Status'])
```

```python
X = dataset.iloc[:,:-1].values
```

**Split the dataset into 90% training and 10% testing**

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,train_size = 0.9,random_state = 10)
```

**Creating the ANN Network**

```python
# Creating an ANN
from keras.models import Sequential
from keras.layers import Dense

# create model
model = Sequential()

# input layer
model.add(Dense(100, input_dim=X.shape[1], activation='sigmoid'))

# Hidden layer 
model.add(Dense(500, activation = 'sigmoid'))
model.add(Dense(500, activation = 'sigmoid'))
model.add(Dense(50,activation = 'sigmoid'))

# Output layer
model.add(Dense(1, activation = 'sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the dataset with 1000 epochs
history = model.fit(X_train,y_train,epochs = 1000,batch_size=1)
```

**Evaluating Training**

```python
import matplotlib.pyplot as plt
losses = history.history['loss']
accuracy = history.history['acc']

plt.plot(history.epoch, accuracy, label = "Accuracy")
plt.plot(history.epoch, losses, label = "Losses", color ='red')
plt.legend()
```

**Evaluating Model**

```python
# Predict the y values
y_pred = model.predict(X_test).astype(int)

# Evaluate the classifier
from sklearn.metrics import accuracy_score, confusion_matrix
print("Accuracy score: ", accuracy_score(y_test, y_pred))
print("Confucion Matrics: \n", confusion_matrix(y_test, y_pred))
```
